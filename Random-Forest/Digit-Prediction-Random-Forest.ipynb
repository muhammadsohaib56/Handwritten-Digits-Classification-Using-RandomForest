```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.gray()
for i in range(10):
    plt.matshow(digits.images[i])
```

```python
dir(digits)
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
digits.target
```

```python
df['target'] = digits.target
df.head()
```

```python
X = df.drop(['target'],axis='columns')
X
```

```python
Y = digits.target
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 10)
```

```python
model.fit(x_train,y_train)
#n_estimators = 100 means it is using 100 ,,depend on test_size also
```

```python
model.score(x_test,y_test)   #more trees score is increasing
```

```python
model.predict(x_test)
```

```python
y_predicted = model.predict(x_test)   #this is truth
```

```python
#we will draw confusion matrix
#Truth in one axis and prediction on other axis
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
